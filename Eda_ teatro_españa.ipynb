```python
import pandas as pd
```

```python
teatro = pd.read_csv(r"C:\Users\elaru\OneDrive\Desktop\proyectos_personales\proyecto_teatro\musicales_vs_sector_2016_2023.csv")
teatro
```

```python
teatro = teatro.rename(columns={"year": "año",
"spectators_theater":"espectadores_teatro","revenue_theater_eur": "ingresos_teatro-eur"})
```

```python
pd.set_option('display.float_format', '{:.0f}'.format)
teatro = teatro.round(0).astype(int)
teatro
```

```python
teatro.describe()
```

```python
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# CONFIGURACIÓN GENERAL
# ---------------------------------------------------------
# 1) Evolución de musicales estrenados por año
# ---------------------------------------------------------
plt.figure(figsize=(8,4))
plt.plot(teatro['año'], teatro['musicales_estrenados'], marker='o')
plt.xlabel('Año')
plt.ylabel('Musicales estrenados')
plt.title('Evolución de musicales estrenados en España')
plt.grid(True)
plt.show()

# ---------------------------------------------------------
# 2) Evolución de espectadores por año
# ---------------------------------------------------------
plt.figure(figsize=(8,4))
plt.plot(teatro['año'], teatro['espectadores_teatro'], marker='o')
plt.xlabel('Año')
plt.ylabel('Espectadores')
plt.title('Evolución de espectadores de teatro en España')
plt.grid(True)
plt.show()

# ---------------------------------------------------------
# 3) Ingresos del teatro por año
# ---------------------------------------------------------
plt.figure(figsize=(8,4))
plt.plot(teatro['año'], teatro['ingresos_teatro-eur'], marker='o')
plt.xlabel('Año')
plt.ylabel('Ingresos (€)')
plt.title('Ingresos anuales del teatro en España')
plt.grid(True)
plt.show()

# ---------------------------------------------------------
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(9,5))

# 1) Musicales en el eje izquierdo
ax1.plot(teatro['año'], teatro['musicales_estrenados'], marker='o', color='tab:blue', label='Musicales')
ax1.set_xlabel('Año')
ax1.set_ylabel('Musicales estrenados', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 2) Espectadores: segundo eje a la derecha
ax2 = ax1.twinx()
ax2.plot(teatro['año'], teatro['espectadores_teatro'], marker='o', color='tab:orange', label='Espectadores')
ax2.set_ylabel('Espectadores', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# 3) Ingresos: tercer eje (derecha desplazada)
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # mueve eje a la derecha
ax3.plot(teatro['año'], teatro['ingresos_teatro-eur'], marker='o', color='tab:green', label='Ingresos €')
ax3.set_ylabel('Ingresos (€)', color='tab:green')
ax3.tick_params(axis='y', labelcolor='tab:green')

plt.title('Comparativa del sector teatral en España')
plt.grid(True)
plt.show()

```
